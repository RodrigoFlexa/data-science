```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
dados.notnull()
```

```python
lista = []
for i in dados.columns:
    lista.append(i.lower())
dados.columns = lista
```

## Tirando os valores nulos 

```python
#Selecionando os valores nulos no preço
dados['valor'].isnull()
```

```python
dados[dados['valor'].isnull()].head(10)
```

```python
dados.dropna(subset = 'valor', inplace = True)
```

```python
dados[dados['valor'].isnull()].head(10)
```

```python
dados.head(20)
```

## Trocando os valores nulos por algum outro 

```python
#O IPTU e o condominio serão tratados como valores 0
```

```python
selecao = (dados['iptu'].isnull()) | (dados['condominio'].isnull())
dados_nulos = len(dados[selecao])
print("dados nulos: ", dados_nulos)
```

```python
dados[selecao]
```

```python
dados.fillna({'condominio': 0, 'iptu': 0},inplace = True)
```

```python
selecao = (dados['iptu'].isnull()) | (dados['condominio'].isnull())
dados_nulos = len(dados[selecao])
print("dados nulos: ", dados_nulos)
```
